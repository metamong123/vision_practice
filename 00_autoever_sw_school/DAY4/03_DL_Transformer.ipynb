### CNN의 한계
- 서로 멀리 떨어진 부분들 사이의 관계를 잘 처리하지 못함
- 멀리 떨어진 feature들을 고려하기 위해서는 큰 필터 필요 -> 계산 효율성 감소

### Recurrent Neural Network(순환 신경망)
- 자연어 처리에서 흔히 사용하는 구조
- 이전 입력이 이후 입력 처리에 영향을 줌
- 먼 거리에 떨어진 입력 간의 정보 전달이 어려움 + vanishing gradient 문제
-> Attention

### Attention Mechanism(주의 매커니즘)
- 이전의 입력 중 현재 입력을 처리하는 데 필요한 정보를 가져오는 방법
- 현재 단계의 값과 이전 단계의 값을 비교하여 attention weight를 계산
- 이전 단계의 값들에 attention weight를 적용하여 가중합
- 현재 단계와 유사한 특성을 갖는 이전 단계의 값들을 더 큰 비중으로 반영

### Transformer
- Attention Mechanism만을 사용한 자연어 처리 모형(RNN 부분을 제거)
- 문장 내/문장 간 attention mechanism 적용

### Vision Transformer
- 이미지를 패치(patch)로 쪼개어, 패치의 시퀀스로 다룸
- Self Attention은 순열 불변(Permutation Invariant) -> 입력 순서가 달라져도 출력이 같음
- 위치 인코딩을 함께 입력 -> 순서 정보 추가
