```python
import cv2
import numpy as np
from PIL import Image

def show(image):
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return Image.fromarray(rgb)
```

```python
# 크기 변경
image = cv2.imread('../data/balls.png')

height, width, channel = image.shape # 원래 크기
new_size = (width//2, height//2) # 절반 크기
image = cv2.resize(image, new_size)
show(image)
```

```python
# 비율로 변경
image = cv2.resize(image, (0, 0), fx=0.5, fy=0.5)
show(image)
```

```python
# 보간
# 2배로 확대
big = cv2.resize(image, None, fx=2.0, fy=2.0, interpolation=cv2.INTER_LANCZOS4)
show(big)
```

```python
# 뒤집기
# 상하
flip_ud = cv2.flip(image, 0)
# 좌우
flip_lr = cv2.flip(image, 1)
# 상하좌우
flip_udlr = cv2.flip(image, -1)

show(np.concatenate([flip_ud, flip_lr, flip_udlr], axis=1))
```

```python
# 회전
# 시계방향 90도
rot_c90 = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
# 180도
rot_180 = cv2.rotate(image, cv2.ROTATE_180)
# 반시계방향 90도
rot_cc90 = cv2.rotate(image, cv2.ROTATE_90_COUNTERCLOCKWISE)

# show(rot_c90)
# show(rot_180)
show(rot_cc90)
```

```python
# 이미지 자르기
height, width, channel = image.shape
x, y = width // 2, height // 3 # 자를 위치(가로 1/2 위치, 세로 1/3 위치)
w, h = width // 4, height // 5 # 자를 크기(가로 1/4 크기, 세로 1/5 크기)
cropped = image[y:y+h, x:x+w] # 이미지 자르기
show(cropped)
```
